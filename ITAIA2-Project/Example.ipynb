```python
# ================================================
# Assignment 1 - Question 1: Complete Example Code
# Delivery Route Optimization using Search Algorithms
# ================================================

import heapq
from collections import defaultdict, deque

# ================================================
# 1.5 Graph Representation (Adjacency List with weights)
# ================================================
graph = {
    'A': {'B': 4, 'C': 2},
    'B': {'D': 5},
    'C': {'D': 8, 'E': 10},
    'D': {'F': 6},
    'E': {'F': 3},
    'F': {}
}

print("Graph representation (1.5):")
for node, neighbors in graph.items():
    print(f"  {node} -> {neighbors}")
print()

# ================================================
# Helper function to reconstruct path from came_from dict
# ================================================
def reconstruct_path(came_from, current):
    path = []
    while current is not None:
        path.append(current)
        current = came_from.get(current)
    return path[::-1]  # reverse to get start -> goal

# ================================================
# 1.6 + 1.7 Uniform Cost Search (UCS) Implementation
# ================================================
def uniform_cost_search(graph, start, goal):
    """
    UCS finds the lowest-cost path from start to goal.
    Uses a priority queue ordered by cumulative path cost.
    """
    # Priority queue: (cost_so_far, node)
    pq = [(0, start)]
    came_from = {start: None}
    cost_so_far = {start: 0}

    while pq:
        current_cost, current = heapq.heappop(pq)

        if current == goal:
            path = reconstruct_path(came_from, goal)
            return path, current_cost

        for neighbor, weight in graph.get(current, {}).items():
            new_cost = current_cost + weight
            
            # Only update if this is a cheaper path to neighbor
            if neighbor not in cost_so_far or new_cost < cost_so_far[neighbor]:
                cost_so_far[neighbor] = new_cost
                came_from[neighbor] = current
                heapq.heappush(pq, (new_cost, neighbor))

    return None, float('inf')  # No path found

print("=== Uniform Cost Search (UCS) ===")
path, total_cost = uniform_cost_search(graph, 'A', 'F')
print(f"Path found: {' -> '.join(path)}")
print(f"Total cost: {total_cost} km\n")

# ================================================
# Bonus: BFS for comparison (helps answer 1.1 & 1.2)
# ================================================
def bfs(graph, start, goal):
    """Breadth-First Search - explores level by level (fewest edges)"""
    queue = deque([(start, [start])])
    visited = set([start])

    while queue:
        current, path = queue.popleft()
        if current == goal:
            return path, len(path) - 1  # number of edges as "cost"
        
        for neighbor in graph.get(current, {}):
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append((neighbor, path + [neighbor]))
    return None, float('inf')

print("=== BFS for comparison (Q1.1 & 1.2) ===")
bfs_path, bfs_cost = bfs(graph, 'A', 'F')
print(f"BFS Path: {' -> '.join(bfs_path)}")
print(f"BFS 'cost' (edges): {bfs_cost}\n")

# ================================================
# 1.8 Constraint Satisfaction Problem (CSP) Formulation
# ================================================
"""
CSP Variables, Domains, and Constraints for Q1.8

Variables:
- Route: sequence of nodes from A to F
- Or more simply: binary variables for key decisions

Simple model using variables:
- Visit_C: Boolean (must visit C?)
- Visit_D: Boolean (must visit D?)
- Total_Distance: integer (must be <= 15)
"""

# Simple backtracking CSP solver example (for learning)
def is_valid_path(path, graph, max_distance=15):
    """Check if path satisfies constraints"""
    if path[0] != 'A' or path[-1] != 'F':
        return False
    
    total = 0
    visits_c_or_d = False
    
    for i in range(len(path)-1):
        u, v = path[i], path[i+1]
        if u not in graph or v not in graph[u]:
            return False
        total += graph[u][v]
        if v in ['C', 'D'] or u in ['C', 'D']:
            visits_c_or_d = True
    
    return visits_c_or_d and total <= max_distance

print("=== CSP Constraints (Q1.8) ===")
print("Variables: nodes visited in route")
print("Domain: valid connections from the graph")
print("Constraints:")
print("  1. Route must start at A and end at F")
print("  2. Must pass through C or D (or both)")
print("  3. Total distance <= 15 km")
print("Example valid path from UCS:", path, "satisfies constraints?", 
      is_valid_path(path, graph))
print()

# ================================================
# Summary Output for Assignment
# ================================================
print("=== SUMMARY FOR YOUR ASSIGNMENT ===")
print("1.1 BFS prioritises fewest hops (not lowest cost)")
print("1.2 UCS uses cumulative cost (priority queue) vs BFS FIFO queue")
print("1.3 Heuristic (e.g. straight-line distance to F) guides search towards goal")
print("1.4 Best algorithm: A* (UCS + heuristic) when heuristic is available")
print("1.5 Graph: adjacency dict (above)")
print("1.6-1.7 UCS implemented and path found")
print("1.8 CSP: variables, domains, constraints defined above")
```
