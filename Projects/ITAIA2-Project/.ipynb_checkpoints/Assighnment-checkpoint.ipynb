# Question 1
## 1.5 Graph

```python
graph = {
    'A': {'B': 4, 'C': 2},
    'B': {'D': 5},
    'C': {'D': 8, 'E': 10},
    'D': {'F': 6},
    'E': {'F': 3},
    'F': {} #goal
}

print("Graph representation 1.5)")
for node, neighbors in graph.items():
    print(f"  {node} -> {neighbors}")
```

## 1.6 & 1.7 UCS

```python
# function to reconstrunct path traverced
def reconstruct_path(came_from, current):
    path = []
    while current is not None:
        path.append(current)
        current = came_from.get(current)
    return path[::-1]

# main UCS function **1.6**
def ucs(graph, start, goal):
    pq = [(0, start)]  # priority que
    came_from = {start: None}
    cost_so_far = {start: 0}

    while pq:
        current_cost, current = heapq.heappop(pq)

        if current == goal:
            path = reconstruct_path(came_from, goal)
            return path, current_cost

        for neighbor, weight in graph.get(current, {}).items():
            new_cost = current_cost + weight
            
            if neighbor not in cost_so_far or new_cost < cost_so_far[neighbor]:
                cost_so_far[neighbor] = new_cost
                came_from[neighbor] = current
                heapq.heappush(pq, (new_cost, neighbor))

    return None, float('inf')  # no path found

# running function
path, total_cost = ucs(graph, 'A', 'F')

# displaying the path & cost **1.7**
print(f"  Path: {' -> '.join(path)}")
print(f"  Cost: {total_cost} km\n")
```

## 1.8 Constraints

```python
# function to test if path is valid
def validate_path(path, graph, max_distance=15):  # sets maximum distance to 15km
    """Check if path satisfies constraints"""
    if path[0] != 'A' or path[-1] != 'F':  # test if path is made
        return False
    
    total = 0
    c_or_d = False
    
    for i in range(len(path)-1):  # walks through path
        u, v = path[i], path[i+1]
        if u not in graph or v not in graph[u]:
            return False
        total += graph[u][v]  # sums distance for total in km
        if v in ['C', 'D'] or u in ['C', 'D']:  # tests for c or d
            c_or_d = True
    
    return c_or_d and total <= max_distance  # only returns true if both are true passed C or D and 15km or under

print("Does path meet condtraints?")
if validate_path(path, graph):
    print("  YES!!!")
else:
    print("  no :(")
```
